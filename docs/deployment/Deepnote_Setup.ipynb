```python
# Step 2: Verify Files
import os

required_files = [
    'deepnote_app.py',
    'agents/customer_success_agent.py',
    'agents/procurement_agent.py',
    'agents/revenue_agent.py',
    'workflows/orchestrator.py',
    'data/customer_success_data.csv'
]

missing = []
for file in required_files:
    if not os.path.exists(file):
        missing.append(file)
    else:
        print(f"‚úÖ {file}")

if missing:
    print(f"\n‚ö†Ô∏è Missing files: {', '.join(missing)}")
    print("Please upload these files to Deepnote.")
else:
    print("\n‚úÖ All required files found!")
```

```python
# Step 3: Start Streamlit App
import subprocess
import sys
import os

print("üöÄ Starting Streamlit app...")
print("=" * 50)
print("\nüìç Look for a URL in the output below!")
print("   Click it to open your dashboard.\n")
print("=" * 50)

# Determine which app file to use
app_file = "deepnote_app.py" if os.path.exists("deepnote_app.py") else "app.py"

# Run Streamlit
subprocess.run([
    sys.executable, "-m", "streamlit", "run",
    app_file,
    "--server.port", "8501",
    "--server.address", "0.0.0.0",
    "--server.headless", "true",
    "--server.enableCORS", "false",
    "--server.enableXsrfProtection", "false"
])
```
